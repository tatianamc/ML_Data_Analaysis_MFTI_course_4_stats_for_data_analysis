Task 1 По данным опроса, 75% работников ресторанов утверждают, что испытывают на работе существенный стресс, оказывающий негативное влияние на их личную жизнь. Крупная ресторанная сеть опрашивает 100 своих работников, чтобы выяснить, отличается ли уровень стресса работников в их ресторанах от среднего. 67 из 100 работников отметили высокий уровень стресса.

Посчитайте достигаемый уровень значимости, округлите ответ до четырёх знаков после десятичной точки.

Переформулируем вопрос в терминах проверки гипотез:"отличается ли уровень стресса сотрудников ресторана от среднего?" 

Имеем дело с выборкой, состоящей из 0 и 1, то есть, из распределения Бернулли  Ber(p) . 

Гипотеза H0- верно ли сети ресторанов распределение ответов сотрудников -биномиальное распределение  Bin(n,p) c p=0.75 ? 

Гипотеза H1: p не равно 0.75. Используем друстороннюю альтернативу, учитываем как правый так и левый конец, так как интересует отклонение доли от среднего.

```python
import numpy as np
from scipy import stats
```

```python
%pylab inline
```

гипотеза  H : Джеймс Бонд предпочитает какой-то определённый вид мартини, при этом он не выбираем какой именно. При расчете будем учитывать как правый, так и левый конец.


```python
stats.binom_test(67, 100, 0.75, alternative = 'two-sided')
```

Если сосны действительно растут равномерно, какое среднее ожидаемое количество сосен в каждом квадрате? В правильном ответе два знака после десятичной точки.


```python
import pandas as pd
data = pd.read_csv("./pines.csv")
data.head(5)
```

```python
s=s.convert_objects(convert_numeric=True)
s0=s0.convert_objects(convert_numeric=True)
```

```python
data.we.reshape(-1,1)
```

```python
our_bins=[0.0,40,80,120,160,200]
```

```python
res=stats.binned_statistic_2d(x=s0, y=s, values=None,statistic='count', bins=[our_bins,our_bins])
```

```python
res.statistic
```

```python
mean(list(map(np.mean, res.statistic)))
```

Чтобы сравнить распределение сосен с равномерным, посчитайте значение статистики хи-квадрат для полученных 5х5 квадратов. Округлите ответ до двух знаков после десятичной точки.

```python
stats.chisquare(res.statistic.reshape(-1,1))
```

Получили значение статистики 150.59, p-value практически ноль. Значит, можем отвергнуть гипотезу H0. То есть наши данные не имеют нормальное распределение.
